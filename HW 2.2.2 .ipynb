```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
```

```python
data = np.genfromtxt(url, delimiter = ',',  dtype=['f4','f4','f4','f4','U16'])
```

```python
data.dtype.names = ('v1','v2','v3','v4', 'group')
```
